```python
import numpy as np
import pandas as pd
```

### Addition of two matrices

```python
A = [[5, 8, 3],
    [6, 4, 9],
    [0, 6, 6]]

B = [[9, 3, 6],
    [4, 6, 2],
    [5, 5, 5]]

result = [[0,0,0],
         [0,0, 0],
         [0, 0, 0]]

```

```python
# Iterate through rows
for i in range (len(A)):
    # iterate through columns
    for j in range (len(A[0])):
        result[i][j] = A[i][j] + B[i][j]

for r in result:
    print(r)
```

```python
student_means = np.mean(result, axis = 1)
student_means
```

```python
np.mean(result)
```

```python
weight = [0.35, 0.40, 0.25]
```

```python
np.dot(weight, result) # by columns. If we reverse the order it is by rows
```
